```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
sistema = [
    dict(m = 1, r = [0, 0, 0], v = [0, 0, 0]),
    dict(m = 1, r = [1, 0, 0], v = [0, 1, 0])
]
```

```python
ts = np.linspace(0, 10, 100)
rs, vs, rps, vps, constantes = pc.ncuerpos_solucion(sistema, ts)
```

```python
plt.plot(rs[0, :, 0], rs[0, :, 1], label='Cuerpo 1')
plt.plot(rs[1, :, 0], rs[1, :, 1], label='Cuerpo 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trayectorias de los cuerpos')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(rps[0, :, 0], rps[0, :, 1], label='Cuerpo 1')
plt.plot(rps[1, :, 0], rps[1, :, 1], label='Cuerpo 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trayectorias de los cuerpos')
plt.legend()
plt.grid()
plt.show()
```

## Cuadraturas

### Momentum lineal total

```python
suma = 0
for i in range(2):
    suma += sistema[i]['m'] * vs[i, 12, :]

suma
```

```python
sistema[0]['m'] * vs[0] + sistema[1]['m'] * vs[1] 
```

### Momentum angular total

```python
suma = 0
for i in range(2):
    suma += sistema[i]['m'] * np.cross(rs[i, 50, :], vs[i, 50, :])

suma
```

```python
# Usamos todo el rango temporal (segundo índice :) para el momento angular total
suma = (sistema[0]['m'] * np.cross(rs[0, :, :], vs[0, :, :]) +
    sistema[1]['m'] * np.cross(rs[1, :, :], vs[1, :, :]))

suma
```
